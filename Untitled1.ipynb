```python
import matplotlib.pyplot as plt
x=[5,2,7]
y=[2,16,4]
plt.plot(x,y)
plt.title("This is title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
y1=[30,89,70,60,80]
y2=[45,67,89,90,56]
y3=[78,90,89,67,89]
plt.plot(x,y,label='green',linewidth=2)
plt.plot(x,y1,label='red',linewidth=3)
plt.plot(x,y2,label='yellow',linewidth=4)
plt.plot(x,y3,label='red',linewidth=5)
plt.title("This is title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```
